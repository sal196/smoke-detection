Featue selection

```python
df
```

```python
x=df.iloc[:,:-1]
y=df["smoking"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

logistic

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

Bagging Algorithim

```python
from sklearn.ensemble import BaggingClassifier
```

```python
B_clf=BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=1000)
```

```python
B_clf.fit(x_train,y_train).score(x_test,y_test)
```

```python
y_pred=B_clf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

Bagging Algorith-Extra trees

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
EX=ExtraTreesClassifier(n_estimators=1000,random_state=42)
```

```python
EX.fit(x_train,y_train)
```

```python
y_pred=EX.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=1000)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```
